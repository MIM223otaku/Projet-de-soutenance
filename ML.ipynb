```python
df = df.drop("ID_Patient",axis =1)
print(df.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
encodeur = LabelEncoder()

df["Sexe"]= encodeur.fit_transform(df["Sexe"])
df["Zone"]=encodeur.fit_transform(df["Zone"])
df["Saison"]=encodeur.fit_transform(df["Saison"])
print(df[["Sexe","Zone","Saison"]].head(10))
```
